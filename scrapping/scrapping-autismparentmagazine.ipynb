This aims to get relevant data from:   
https://www.autismparentingmagazine.com/applied-behavior-analysis-aba/

```python
import requests  # to make GET request
from bs4 import BeautifulSoup  # to parse the HTML response
import time  # to pause between calls
import pandas as pd  # to see CSV
```

```python
def parse_page(response):
    global df
    global source
    # parse the response into an HTML tree soup object
    # read the content of the server’s response as a string
    page_source = response.text
    soup = BeautifulSoup(page_source, 'html5lib')
    # take a look
    #print(soup.prettify()[:1000])

    # Find all "<div class="entry-row" appearances
    links=soup.select("div.entry-row")
    for link in links:
        article=link.select("article")
        # Get category (for validation)
        for element in article[0]['class']:
            if ( 'category' in element ):
                category=element
                break
                
        # Get text
        #article[0].find_all('div')[1].get_text()
        text=article[0].find('div', class_='entry-summary entry-content').get_text()
    
        # Get href. for further reading
        href=article[0].find('a')['href']
    
    
        newrow={"source":source,"category":category,"text":text,"href":href}

        # Add entry into database
        df.loc[len(df.values)]=newrow
```

```python
# Create an empty database to keep the history of purchases
columns=["source","category","text","href"]
df=pd.DataFrame(columns=columns)
```

```python
source="https://www.autismparentingmagazine.com/"
    
# Get list of pages to parse:
f = open("data/AutismParentMagazine-urls.txt","r")
urls =  f.readlines()
f.close()

for ii in range(len(urls)):
    url = urls[ii]
    urls[ii]=url.strip()
```

```python
#loop over urls:

for url in urls:
    # make a GET request
    response = requests.get(url)

    parse_page(response)
    time.sleep(5)
```

```python
# Convert to file
df.to_csv("data/AutismParentMagazine-posts.csv")
```
